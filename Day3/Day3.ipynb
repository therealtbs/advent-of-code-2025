# Advent of Code - Day 3
[Day 3: Lobby](https://adventofcode.com/2025/day/3)

```kotlin
%use dataframe
val input = DataFrame.readCsv("./input.txt", header = listOf("input"), colTypes = mapOf("input" to ColType.String))
input
```

## Part 1

```kotlin
val digitmap = input.map { it.input }.associateWith {
    it.mapIndexed { i, char -> Pair(i, char.digitToInt()) }.sortedByDescending { (_, digit) -> digit }
}

val maxJoltage = digitmap.mapValues { (_, sortedDigits) ->
    val (firstIndex, firstDigit) = sortedDigits.first { (searchIndex, _) -> sortedDigits.any { (index, _) -> index > searchIndex } }
    val (_, secondDigit) = sortedDigits.first { (index, _) -> index > firstIndex }

    (firstDigit*10)+secondDigit
}

DISPLAY(digitmap.toDataFrame())
maxJoltage
```
