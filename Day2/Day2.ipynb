# Advent of Code 2025 - Day 2
[Day 2: Gift Shop](https://adventofcode.com/2025/day/2)

```kotlin
%use dataframe
val input = DataFrame.read("./input.txt")
val ranges = input.columns().map {it.name().split('-').map { it.toULong() }}
```

```kotlin
val invalidIdsPerRange = ranges.flatMap {
    (start, end) -> start.rangeTo(end).filter {
        val idString = it.toString()
        // If uneven number of digits, bail -> cannot not invalid
        if (idString.length % 2 != 0) return@filter false
        // Split string down the middle and compare
        val (firstHalf, secondHalf) = idString.chunked(idString.length / 2)
        firstHalf == secondHalf
    }
}
invalidIdsPerRange.toDataFrame()
```

```kotlin
invalidIdsPerRange.sum()
```
